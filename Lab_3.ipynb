```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Assuming 'df' is the initial DataFrame
scaler = StandardScaler()

# Apply Z-score normalization to 'Alcohol' and 'Malic acid'
df[['Alcohol', 'Malic acid']] = scaler.fit_transform(df[['Alcohol', 'Malic acid']])

# Plot the density graph for normalized Malic acid
sns.kdeplot(df['Malic acid'], label='Malic acid')
plt.title('Density plot of Normalized Malic acid')
plt.xlabel('Z-score Normalized Malic acid')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'df' is the initial DataFrame

# Apply log transformation to 'Total phenols' and 'Flavanoids'
df['Log Total phenols'] = np.log(df['Total phenols'] + 1)  # Adding 1 to avoid log(0)
df['Log Flavanoids'] = np.log(df['Flavanoids'] + 1)

# Plot the distribution for 'Total phenols'
plt.figure(figsize=(10, 5))

# Plot for Log Total phenols
plt.subplot(1, 2, 1)
sns.kdeplot(df['Log Total phenols'], label='Total phenols')
plt.title('Log Transformed Total Phenols')
plt.xlabel('Log Total Phenols')
plt.ylabel('Density')
plt.legend()

# Plot for Log Flavanoids
plt.subplot(1, 2, 2)
sns.kdeplot(df['Log Flavanoids'], label='Flavanoids')
plt.title('Log Transformed Flavanoids')
plt.xlabel('Log Flavanoids')
plt.ylabel('Density')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset (replace 'your_file.csv' with the actual file path)
# df = pd.read_csv('your_file.csv')

# Assuming 'wine_class' is the column name for the wine class attribute, drop it
df = df.drop(columns=['wine_class'])

# Normalize all other columns using Z-score normalization
df_normalized = (df - df.mean()) / df.std()

# Calculate the correlation matrix
corr_matrix = df_normalized.corr()

# Display the correlation matrix in numbers
print(corr_matrix)

# Plot the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='viridis')
plt.title('Correlation Matrix')
plt.show()

# Find pairs of features with correlation higher than 0.8
high_corr_pairs = np.where(np.abs(corr_matrix) > 0.8)
high_corr_pairs = [(corr_matrix.index[x], corr_matrix.columns[y])
                   for x, y in zip(*high_corr_pairs) if x != y and x < y]

# Display highly correlated pairs
print("Highly correlated pairs (correlation > 0.8):")
for pair in high_corr_pairs:
    print(pair)
```
